```python
import torch
import torch.nn as nn
import torch.optim as optim

X = torch.tensor([
    [[1.],[2.],[3.]],
    [[2.],[3.],[4.]],
    [[3.],[4.],[5.]],
    [[7.],[8.],[9.]],
    [[8.],[9.],[10.]],
    [[9.],[10.],[11.]]
], dtype=torch.float32)

y = torch.tensor([0,0,0,1,1,1], dtype=torch.float32).view(-1,1)

class GRUModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.gru = nn.GRU(input_size=1, hidden_size=16, batch_first=True)
        self.fc = nn.Linear(16,1)

    def forward(self, x):
        out, _ = self.gru(x)
        out = out[:,-1,:]
        out = self.fc(out)
        out = torch.sigmoid(out)
        return out

model = GRUModel()

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(200):
    output = model(X)
    loss = criterion(output, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

test = torch.tensor([[[4.],[5.],[6.]]], dtype=torch.float32)

with torch.no_grad():
    prediction = model(test)

print(prediction)
print("Class:", 1 if prediction.item() > 0.5 else 0)
```
